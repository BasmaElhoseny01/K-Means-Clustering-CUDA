```python
# read image to cluster
img = plt.imread('../tests/high_res.jpg')
img = img / 255
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
kmeansRnd = KMeans(n_clusters=5, init='random',random_state=42, max_iter=100, n_init=1)
```

```python
# reshape image to 2D array
pixels = img.reshape(-1, 3)

start_time = time.time()
# fit and predict
kmeansRnd.fit(pixels)
end_time = time.time()
print(end_time - start_time)
```

```python
# show the image results
kmeansRnd_img = kmeansRnd.cluster_centers_[kmeansRnd.labels_]
kmeansRnd_img = kmeansRnd_img.reshape(img.shape)
plt.imshow(kmeansRnd_img)
plt.axis('off')
plt.show()
```
